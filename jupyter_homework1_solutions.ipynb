## The Pythagorean Theorem

In a right triangle, the square of the hypotenuse equals the sum of the squares of the other two sides:

$$a^2 + b^2 = c^2$$

Where:
- $a$ and $b$ are the lengths of the two legs
- $c$ is the length of the hypotenuse

---
### Explanation for Question 4:
- `$$formula$$` creates a centered, block-level mathematical formula (LaTeX)
- `$variable$` creates inline math notation
- `^2` creates a superscript (squared) in LaTeX
- Jupyter supports LaTeX for mathematical notation

```python
# QUESTION5
# This exercise demonstrates variable creation, arithmetic operations, and type checking

# Create variable x with integer value 3
# Integers are whole numbers without decimal points
x = 3

# Create variable y with float value 3.141 (approximation of pi)
# Floats are numbers with decimal points
y = 3.141

# Calculate AREA using the formula y * x^2
# In Python, ** is the exponentiation operator (power)
# So x**2 means x squared (x to the power of 2)
AREA = y * x ** 2

# Print the values of all variables
# f-strings (f"...") allow us to embed variables directly in strings
print(f"x = {x}")
print(f"y = {y}")
print(f"AREA = {AREA}")

# Print the types of each variable using type() function
# type() returns the data type of a variable
print(f"\nType of x: {type(x)}")      # Expected: <class 'int'>
print(f"Type of y: {type(y)}")        # Expected: <class 'float'>
print(f"Type of AREA: {type(AREA)}")  # Expected: <class 'float'>
```

```python
# QUESTION6
# This exercise demonstrates comparison and logical operators
# We use the AREA variable from the previous cell

# First, let's see what AREA value we're working with
print(f"AREA value: {AREA}")
print(f"AREA // 4 = {AREA // 4}")  # Floor division (integer result)
print(f"AREA / 4 = {AREA / 4}")    # Regular division (float result)
print()

# Expression 1: AREA // 4 <= 7
# // is floor division - divides and rounds down to nearest integer
# <= is less than or equal to comparison
result1 = AREA // 4 <= 7
print(f"AREA // 4 <= 7 : {result1}")

# Expression 2: AREA / 4 < 7
# / is regular division (returns float)
# < is less than comparison
result2 = AREA / 4 < 7
print(f"AREA / 4 < 7 : {result2}")

# Expression 3: not AREA / 4 < 7 and AREA // 4 <= 7
# 'not' negates the boolean value (True becomes False, False becomes True)
# 'and' returns True only if BOTH conditions are True
# Order of operations: 'not' is evaluated first, then 'and'
result3 = not AREA / 4 < 7 and AREA // 4 <= 7
print(f"not AREA / 4 < 7 and AREA // 4 <= 7 : {result3}")
```

```python
# QUESTION7
# This exercise demonstrates string indexing and slicing

# Create a string with full name (first and last name separated by space)
str_name = "Tomer Doe"

print(f"Full string: '{str_name}'")
print(f"String length: {len(str_name)}")
print()

# Print all values at EVEN indexes (0, 2, 4, 6, ...)
# String indexing: string[start:stop:step]
# [::2] means start from beginning, go to end, step by 2
# Index:  0   1   2   3   4   5   6   7   8
# Char:   T   o   m   e   r   ' '  D   o   e
even_indexes = str_name[::2]
print(f"Characters at even indexes (0,2,4,...): '{even_indexes}'")

# Print all values at ODD indexes (1, 3, 5, 7, ...)
# [1::2] means start from index 1, go to end, step by 2
odd_indexes = str_name[1::2]
print(f"Characters at odd indexes (1,3,5,...): '{odd_indexes}'")
```

```python
# QUESTION8 (Challenge)
# This exercise demonstrates string concatenation, the find() method, and f-strings

# Create variable 'a' with 5 hash symbols
a = "#####"

# Create variable 'b' with 5 asterisks
b = "*****"

# Create variable 'c' with your name
c = "Tomer Doe"

# Concatenate variables in order: a, c, b
# The + operator joins strings together
concatenated = a + c + b
print(f"Concatenated string: '{concatenated}'")

# Find the index of the space character using .find() method
# .find() returns the index of the first occurrence of the specified value
# Note: We search in 'c' since that's where the name with space is
space_index_in_c = c.find(" ")
print(f"Index of space in name: {space_index_in_c}")

# To find space in concatenated string, we need to account for 'a' prefix
space_index_in_concat = concatenated.find(" ")
print(f"Index of space in concatenated string: {space_index_in_concat}")

# Calculate lengths of first and last name
# Split the name at the space to get first and last name separately
# .split() divides a string into a list based on a delimiter
name_parts = c.split(" ")
first_name = name_parts[0]
last_name = name_parts[1]

# Print using f-string the length of first and last name
# len() returns the number of characters in a string
print(f"First name '{first_name}' has {len(first_name)} characters")
print(f"Last name '{last_name}' has {len(last_name)} characters")
```

```python
# QUESTION9 (Challenge)
# This exercise demonstrates using matplotlib to draw circles

# Import matplotlib for plotting
# matplotlib.pyplot is the standard plotting module
import matplotlib.pyplot as plt

# Create a figure and axis object
# fig is the entire figure, ax is the plotting area
fig, ax = plt.subplots()

# Create two red circles that touch each other
# Circle parameters: (x_center, y_center), radius
# For circles to touch, the distance between centers should equal sum of radii

# First circle: centered at (1, 1) with radius 1
circle1 = plt.Circle((1, 1), 1, color='red', fill=False)

# Second circle: centered at (3, 1) with radius 1
# Distance between centers (1,1) and (3,1) is 2
# Sum of radii is 1 + 1 = 2
# Since distance = sum of radii, the circles touch at exactly one point
circle2 = plt.Circle((3, 1), 1, color='red', fill=False)

# Add circles to the axes
# add_patch() adds a shape to the plot
ax.add_patch(circle1)
ax.add_patch(circle2)

# Set axis limits to properly display the circles
# Without this, the circles might be cut off or not visible
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 3)

# Set equal aspect ratio so circles appear as circles (not ellipses)
ax.set_aspect('equal')

# Add title for clarity
plt.title('Two Red Circles Touching Each Other')

# Display the plot
plt.show()
```

---
### Explanation for Question 9:
- `plt.Circle((x, y), radius, color='red')` creates a circle object
- `fill=False` makes the circle hollow (just the outline)
- For two circles to touch exactly, the distance between their centers must equal the sum of their radii
- `ax.add_patch()` adds the circle to the plot
- `ax.set_aspect('equal')` ensures circles look round, not stretched
